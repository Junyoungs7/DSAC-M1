```python
import requests
import json
import pandas as pd
from pandas.io.json import json_normalize
```

```python
data = {
    'ins_lat':'37.56682', # 지정한 위도
    'ins_lng':'126.97865',# 경도에서 가까운 순으로 나열
    'p_sido_cd':'01', # 시도 코드, 01=서울시, 08=경기 ... 16=제주
    'p_gugun_cd':'',  # 구군 코드, 세부지역 (지정하지 않으면 시/도 전체)
    'in_biz_cd': '',
    'set_date': '',
    'iend': 1000,
}

url = 'https://www.starbucks.co.kr/store/getStore.do'  # store-> 매장 찾기-> https://www.starbucks.co.kr/store/getStore.do===> ?disp=locale
r = requests.post(url, data=data)
r.text[:1000] # 수신된 데이터의 앞부분만 확인
```

```python
jo = json.loads(r.text)
df = json_normalize(jo, 'list')
```

```python
len(df)
```

```python
print(df)
```

```python
df.columns
```

```python
df = df[['s_name', 'lat', 'lot', 'sido_name', 'gugun_name', 'doro_address', 'tel']]
df
```

```python
df.dtypes
```

```python
df['lat'] = df['lat'].astype(float)
df['lot'] = df['lot'].astype(float)
df.dtypes
```
